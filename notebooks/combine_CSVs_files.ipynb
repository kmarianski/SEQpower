```python
import pandas as pd
import os
```

```python
CSVs = [f for f in os.listdir("../outputs/") if f.endswith("csv") and "loci" not in f] 
```

```python
df = pd.DataFrame()
for csv in CSVs:
    df_temp = pd.read_csv(f"../outputs/{csv}")
    df = pd.concat([df, df_temp], ignore_index=True)
print(df.shape)
df.head()
```

```python
import os
import shutil
from datetime import datetime

output_dir = "out_csv"
archive_dir = os.path.join(output_dir, "archive")
file_name = "combined_SEQpower.csv"
file_path = os.path.join(output_dir, file_name)

os.makedirs(archive_dir, exist_ok=True)

if os.path.exists(file_path):
    current_date = datetime.now().strftime("%Y-%m-%d")
    dated_archive_dir = os.path.join(archive_dir, current_date)
    os.makedirs(dated_archive_dir, exist_ok=True)
    shutil.move(file_path, os.path.join(dated_archive_dir, file_name))
```

```python
df.to_csv("out_csv/combined_SEQpower.csv", index=False)
```
